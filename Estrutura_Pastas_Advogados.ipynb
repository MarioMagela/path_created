```python
from google.colab import drive
import os

# Montar o Google Drive
drive.mount('/content/drive')

# Função para criar pastas e subpastas
def create_folder_structure(base_path, structure):
    for main_folder, subfolders in structure.items():
        # Cria o diretório principal
        main_folder_path = os.path.join(base_path, main_folder)
        os.makedirs(main_folder_path, exist_ok=True)
        print(f"Criada pasta: {main_folder_path}")

        # Cria os subdiretórios
        for subfolder in subfolders:
            subfolder_path = os.path.join(main_folder_path, subfolder)
            os.makedirs(subfolder_path, exist_ok=True)
            print(f"Criada subpasta: {subfolder_path}")

# Estrutura de pastas corrigida
folder_structure = {
    "ED.01. ADMINISTRAÇÃO": [
        "ED.01.01. Financeiro",
        "ED.01.02. Contabilidade",
        "ED.01.03. Recursos Humanos",
        "ED.01.04. Contratos com Terceiros",
        "ED.01.05. Documentos Legais",
        "ED.01.06. Infraestrutura",
        "ED.01.07. Gestão de Processos Internos"
    ],
    "ED.02. MARKETING": [
        "ED.02.01. Planejamento Estratégico",
        "ED.02.02. Campanhas Publicitárias",
        "ED.02.03. Materiais de Divulgação",
        "ED.02.04. Redes Sociais",
        "ED.02.05. Parcerias e Eventos",
        "ED.02.06. Branding"
    ],
    "ED.03. VENDAS": [
        "ED.03.01. Prospecção de Clientes",
        "ED.03.02. CRM e Leads",
        "ED.03.03. Propostas Comerciais",
        "ED.03.04. Relatórios de Vendas"
    ],
    "ED.04. CLIENTES": [
        "ED.04.01. Documentação de Clientes",
        "ED.04.02. Contratos de Clientes",
        "ED.04.03. Atualização de Status de Processos",
        "ED.04.04. Reuniões e Correspondências",
        "ED.04.05. Sucesso do Cliente",
        "ED.04.06. Histórias"
    ],
    "ED.05. JURÍDICO": [
        "ED.05.01. Casos em Andamento",
        "ED.05.02. Casos Encerrados",
        "ED.05.03. Jurisprudência",
        "ED.05.04. Pesquisas Jurídicas",
        "ED.05.05. Modelos de Documentos",
        "ED.05.06. Pareceres e Relatórios"
    ],
    "ED.06. COMPLIANCE": [
        "ED.06.01. Código de Conduta",
        "ED.06.02. Políticas Internas",
        "ED.06.03. Relatórios de Conformidade"
    ],
    "ED.07. TI": [
        "ED.07.01. Infraestrutura Tecnológica",
        "ED.07.02. Segurança da Informação",
        "ED.07.03. Manutenção de Equipamentos",
        "ED.07.04. Softwares e Licenças"
    ],
    "ED.08. TREINAMENTO": [
        "ED.08.01. Treinamentos Internos",
        "ED.08.02. Desenvolvimento Profissional",
        "ED.08.03. Capacitação em Softwares"
    ],
    "ED.09. REUNIÕES": [
        "ED.09.01. Atas de Reuniões",
        "ED.09.02. Pautas e Agendas",
        "ED.09.03. Relatórios de Progresso"
    ],
    "ED.10. PROJETOS ESPECIAIS": [
        "ED.10.01. Parcerias Estratégicas",
        "ED.10.02. Expansão de Serviços",
        "ED.10.03. Novos Mercados"
    ]
}

# Define o caminho base no Google Drive
base_path = "/content/drive/MyDrive/ESTRUTURA"

# Cria a estrutura de pastas
print("Criando estrutura de pastas no Google Drive:")
create_folder_structure(base_path, folder_structure)

# Opcional: Compactar a estrutura em um arquivo zip
import shutil
zip_path = "/content/drive/MyDrive/ESTRUTURA"
shutil.make_archive(zip_path, "zip", base_path)
print(f"\nEstrutura compactada em {zip_path}.zip")

print("\nEstrutura de pastas criada com sucesso! Verifique em /content/drive/MyDrive/ESTRUTURA")
```
